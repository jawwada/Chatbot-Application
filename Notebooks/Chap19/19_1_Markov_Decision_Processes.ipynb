```python
def markov_decision_process_step_deterministic(state, transition_probabilities_given_action, reward_structure, policy):
  """
  Execute one step of a Markov Decision Process with deterministic policy.

  Args:
      state: Current state index
      transition_probabilities_given_action: 3D array P(s'|s,a)
      reward_structure: Array indicating reward values for each state
      policy: Array mapping states to actions

  Returns:
      new_state: New state index based on current state and policy action
      reward: Reward value for the new state
  """
  # Get the action for the current state from the policy
  action = policy[state]

  # Get transition probabilities for current state and action
  state_action_probs = transition_probabilities_given_action[:,state, action]

  # Sample the next state based on transition probabilities
  new_state = np.random.choice(len(state_action_probs), p=state_action_probs)

  # Calculate reward for the new state
  reward = reward_structure[new_state, 0]

  return new_state, reward

```

```python
def markov_decision_process_step_stochastic(state, transition_probabilities_given_action, reward_structure, stochastic_policy):
  """
  Execute one step of a Markov Decision Process with stochastic policy.

  Args:
      state: Current state index
      transition_probabilities_given_action: 3D array P(s'|s,a)
      reward_structure: Array indicating reward values for each state
      stochastic_policy: 2D array P(a|s) mapping states to action probabilities

  Returns:
      new_state: New state index based on current state and sampled action
      reward: Reward value for the new state
      action: Sampled action from the stochastic policy
  """
  # Get action probabilities for the current state
  action_probs = stochastic_policy[:, state]

  # Sample an action from the stochastic policy
  action = np.random.choice(len(action_probs), p=action_probs)

  # Get transition probabilities for current state and sampled action
  state_action_probs = transition_probabilities_given_action[:, state, action]

  # Sample the next state based on transition probabilities
  new_state = np.random.choice(len(state_action_probs), p=state_action_probs)

  # Calculate reward for the new state
  reward = reward_structure[new_state, 0]

  return new_state, reward, action
```

```python
# Set up the reward structure so it matches figure 19.2
reward_structure = np.zeros((16,1))
reward_structure[15] = 1

# Initialize random number seed
np.random.seed(0)
T = 20
# Set up the states, so the fish are in the same positions as figure 19.5
states = np.zeros(T, dtype='uint8')
rewards = np.zeros(T, dtype='uint8')
actions = np.zeros(T-1, dtype='uint8')

states[0] = 0
for t in range(T-1):
  states[t+1],rewards[t+1],actions[t] = markov_decision_process_step_stochastic(states[t], transition_probabilities_given_action, reward_structure, stochastic_policy)

print("Actions", actions)
print("Your States:", states)
print("Your Rewards:", rewards)

mdp_drawer = DrawMDP(n_rows, n_cols)
mdp_drawer.draw(layout, state = states[0], path1=states, policy = stochastic_policy, draw_state_index = True)
```
